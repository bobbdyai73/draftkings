```python
import pandas as pd
import numpy as np
```

```python
def read_concat(csv1, csv2):
    data=pd.read_csv(csv1)
    data2=pd.read_csv(csv2)
    df = pd.concat([data, data2])
    df.reset_index(inplace=True, drop=True)
    
    return df
    
    
    
```
